```python
import pandas as pd
```

```python
df_nac = pd.read_excel('nacimientos/nacimientos_agrupadas.xlsx')
df_def = pd.read_excel('defunciones/defunciones_agrupadas.xlsx')
```

```python
#NACIMIENTOS
from prophet import Prophet

lista_municipios = df_nac['municipio'].unique()

todos_los_pronosticos = []

for municipio_id in lista_municipios:
    print(f"Procesando pronóstico para el Municipio {municipio_id}...")
    
    df_municipio = df_nac[df_nac['municipio'] == municipio_id].copy()
    
    if len(df_municipio) > 1:
        df_prophet = df_municipio[['año', 'total']].rename(columns={
            'año': 'ds',
            'total': 'y'
        })
        df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y')
        
        modelo_prophet = Prophet()
        modelo_prophet.fit(df_prophet)
        
        futuro = modelo_prophet.make_future_dataframe(periods=5, freq='Y')
        pronostico = modelo_prophet.predict(futuro)
        
        pronostico['municipio'] = municipio_id
        
        todos_los_pronosticos.append(pronostico)

df_nac_prophet = pd.concat(todos_los_pronosticos)

print("\n--- PRONÓSTICOS FINALES CON PROPHET ---")
print(df_nac_prophet.tail(10))
```

```python
from prophet import Prophet
from matplotlib import pyplot as plt

lista_municipios = df_def['municipio'].unique()

todos_los_pronosticos = []

for municipio_id in lista_municipios:
    print(f"Procesando pronóstico para el Municipio {municipio_id}...")
    
    df_municipio = df_def[df_def['municipio'] == municipio_id].copy()
    
    if len(df_municipio) > 1:
        df_prophet = df_municipio[['año_defuncion', 'total']].rename(columns={
            'año_defuncion': 'ds',
            'total': 'y'
        })
        df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y')
        
        modelo_prophet = Prophet()
        modelo_prophet.fit(df_prophet)
        
        futuro = modelo_prophet.make_future_dataframe(periods=5, freq='Y')
        pronostico = modelo_prophet.predict(futuro)
        
        pronostico['municipio'] = municipio_id
        
        todos_los_pronosticos.append(pronostico)

df_def_prophet = pd.concat(todos_los_pronosticos)

print("\n--- PRONÓSTICOS FINALES CON PROPHET ---")
print(df_def_prophet.tail(10))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

fig, ax = plt.subplots(figsize=(15, 9))

lista_municipios = df_nac['municipio'].unique()
palette = sns.color_palette("viridis", n_colors=len(lista_municipios))

for i, municipio_id in enumerate(lista_municipios):
    color = palette[i]
    
    hist_data = df_nac[df_nac['municipio'] == municipio_id]
    forecast_data = df_nac_prophet[df_nac_prophet['municipio'] == municipio_id]
    
    label_nac = 'Nacimientos' if i == 0 else None
    ax.plot(hist_data['año'], hist_data['total'], color='cornflowerblue', alpha=0.7, label=label_nac)
    ax.plot(forecast_data['ds'], forecast_data['yhat'], color='mediumblue', linestyle='--')
    ax.fill_between(forecast_data['ds'], forecast_data['yhat_lower'], forecast_data['yhat_upper'],
                    color='cornflowerblue', alpha=0.1)

lista_municipios_def = df_def['municipio'].unique()
for i, municipio_id in enumerate(lista_municipios_def):
    
    hist_data = df_def[df_def['municipio'] == municipio_id]
    forecast_data = df_def_prophet[df_def_prophet['municipio'] == municipio_id]

    label_def = 'Defunciones' if i == 0 else None
    ax.plot(hist_data['año_defuncion'], hist_data['total'], color='salmon', alpha=0.7, label=label_def)
    ax.plot(forecast_data['ds'], forecast_data['yhat'], color='firebrick', linestyle='--')
    ax.fill_between(forecast_data['ds'], forecast_data['yhat_lower'], forecast_data['yhat_upper'],
                    color='salmon', alpha=0.1)

ax.set_xlim(pd.to_datetime('2018-01-01'), pd.to_datetime('2030-01-01'))
ax.set_title('Comparación de Tendencias por Municipio: Nacimientos vs. Defunciones', fontsize=18)
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Total', fontsize=14)
ax.legend()
plt.grid(True)
plt.savefig("vs")
plt.show()
```
